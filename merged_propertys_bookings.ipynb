```python
#Importar librerías necesarias
import pandas as pd
```

```python
# Cargar el archivo Bookings.csv
properties_df = pd.read_csv("dataset/Properties.csv")  # Reemplaza con la ruta si es necesario
bookings_df = pd.read_csv("dataset/Bookings.csv")
properties_df.head(), bookings_df.head()  # Mostrar las primeras filas de ambos datasets
```

```python
#Limpieza y transformacion en el dataset Properties

# Convertir columna de fecha a tipo datetime
properties_df['ReadyDate'] = pd.to_datetime(properties_df['ReadyDate'], errors='coerce')

#transformación en Properties 
properties_df['PropertyType'] = properties_df['PropertyType'].fillna('Unknown')

# Verificar cambios
properties_df.head()
```

```python
# Limpieza y transformacion en el dataset Bookings

# Convertir columnas de fecha a tipo datetime
bookings_df['BookingCreatedDate'] = pd.to_datetime(bookings_df['BookingCreatedDate'], errors='coerce')
bookings_df['ArrivalDate'] = pd.to_datetime(bookings_df['ArrivalDate'], errors='coerce')
bookings_df['DepartureDate'] = pd.to_datetime(bookings_df['DepartureDate'], errors='coerce')

# Manejo de valores nulos en columnas importantes
bookings_df['Channel'] = bookings_df['Channel'].fillna('Unknown')
bookings_df['RoomRate'] = bookings_df['RoomRate'].fillna(0)
bookings_df['Revenue'] = bookings_df['Revenue'].fillna(0)
bookings_df['ADR'] = bookings_df['ADR'].fillna(bookings_df['RoomRate'])
bookings_df['TouristTax'] = bookings_df['TouristTax'].fillna(0)
bookings_df['TotalPaid'] = bookings_df['TotalPaid'].fillna(bookings_df['Revenue'] + bookings_df['TouristTax'])

# Verificar cambios
bookings_df.head()
```

```python
# Analisis Exploratorio de Datos (EDA) en Properties

print("Estadísticas descriptivas de Properties:")
properties_df.describe(include='all')
```

```python
# Celda 6: Analisis Exploratorio de Datos (EDA) en Bookings

print("Estadísticas descriptivas de Bookings:")
bookings_df.describe(include='all')
```

```python
# Celda 7: Unificacion de los datasets usando PropertyId como clave

# Merge de los datasets
merged_df = pd.merge(bookings_df, properties_df, on='PropertyId', how='left')
merged_df.head()
```

```python
# Verificacion de valores nulos en el dataset unificado

print("Valores nulos en el dataset unificado:")
print(merged_df.isnull().sum())

# Opcional: Rellenar valores nulos remanentes después del merge
merged_df.fillna({'PropertyType': 'Unknown', 'Channel': 'Unknown'}, inplace=True)
```

```python
# Guardar el dataset unificado en un archivo CSV listo para analisis

merged_df.to_csv("merged_properties_bookings.csv", index=False)
print("El archivo unificado se ha guardado como merged_properties_bookings.csv")
```
